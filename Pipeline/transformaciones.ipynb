```python
import numpy as np
```

```python
# Calcular el promedio de los números en la lista de 'rating_list'
data['avg_rating_list'] = data['rating_list'].apply(lambda x: np.mean(x) if len(x) > 0 else 0)

# Calcular el porcentaje de satisfacción
data['percent_good_reviews'] = data['labels'].apply(lambda x: np.sum(x == 'POSITIVE') / np.sum((x == 'POSITIVE') | (x == 'NEGATIVE')) if ('POSITIVE' in x) or ('NEGATIVE' in x) else None)
```

```python
# Define una función para asignar categorías
def categorizar_satisfaccion(satisfaccion):
    if satisfaccion < 0.4:
        return 'Very Dissatisfied 😞'
    elif 0.40 <= satisfaccion < 0.6:
        return 'Dissatisfied 😕'
    elif 0.60 <= satisfaccion < 0.9:
        return 'Satisfied 😊'
    else:
        return 'Very Satisfied 😃'
```

```python
# Aplicar la función a la columna "Satisfaccion" y crear una nueva columna "Categoria"
data['satisfaction'] = data['percent_good_reviews'].apply(categorizar_satisfaccion)
```

```python
# Convierte cada lista de fechas a tipo datetime y guarda el resultado en una nueva columna
data['date_list'] = data['date_list'].apply(lambda x: [pd.to_datetime(date) for date in x])

# Calcula la fecha más antigua en cada lista y guárdala en 'earliest_review'
data['earliest_review'] = data['date_list'].apply(lambda x: min(x))

# Calcula la fecha más actual en cada lista y guárdala en 'latest_review'
data['latest_review'] = data['date_list'].apply(lambda x: max(x))
```
