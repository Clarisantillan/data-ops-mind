```python
import pandas as pd
```

```python
parquet_file = r'C:\Users\Rossella\Escritorio\PROYECTO-FINAL-GRUPAL\illinois_dashboard_reco_date.parquet'
```

```python
df = pd.read_parquet(parquet_file, engine = 'auto')
```

```python
df
```

```python
df.info()
```

```python
columnas_deseadas = [
    'identificador',
    'name',
    'address',
    'latitude',
    'longitude',
    'avg_rating',
    'source',
    'line',
    'satisfaction',
    'percent_good_reviews'
]

# Creo un nuevo DataFrame con las columnas seleccionadas
nuevo_df = df[columnas_deseadas]

nuevo_df
```

```python
nuevo_df.describe()
```

```python
## Hago un anÃ¡lisis de posibles outliers con los grÃ¡ficos de caja

import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot para stars
sns.boxplot(x=nuevo_df['avg_rating'])
plt.xlabel('avg_rating')
plt.title('Boxplot de avg_rating')
plt.show()

# Boxplot para review_count
sns.boxplot(x=nuevo_df['percent_good_reviews'])
plt.xlabel('percent_good_reviews')
plt.title('Boxplot de percent_good_reviews')
plt.show()
```

```python
## Realizamos el histograma con bins personalizados
import numpy as np
sns.histplot(data=nuevo_df, x= 'avg_rating', bins= np.arange(0,25,1))
plt.xlabel('avg_rating')
plt.title('Histograma de avg_rating')
plt.show()
```

```python
## Realizamos el histograma con bins personalizados
import numpy as np
sns.histplot(data=nuevo_df, x= 'percent_good_reviews', bins= np.arange(0,5,0.2))
plt.xlabel('percent_good_reviews')
plt.title('Histograma de percent_good_reviews')
plt.show()
```

```python
# Hago un diagrama de dispesiÃ³n entre dos columnas numÃ©ricas
sns.scatterplot(x='percent_good_reviews', y='avg_rating', data=nuevo_df)
plt.title('Diagrama de DispersiÃ³n: percent_good_reviews vs avg_rating')
plt.xlabel('percent_good_reviews')
plt.ylabel('avg_rating')
plt.show()
```

```python
# Creo una matriz de correlaciÃ³n para las columnas numÃ©ricas
correlation_matrix = nuevo_df.corr()

# Visualizamos la matriz de correlaciÃ³n con un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de CorrelaciÃ³n')
plt.show()
```

```python
# Diagrama de barras para una columna line
plt.figure(figsize=(10, 6))
sns.countplot(x='line', data=nuevo_df)
plt.title('Porcentaje_reviews por rubro')
plt.xlabel('Rubro')
plt.ylabel('Porcentaje_reviews')
plt.xticks(rotation=90)
plt.show()
```

```python
# Encuentro el rubro con la mayor cantidad de negocios
max_line = nuevo_df["line"].value_counts().idxmax()

# Filtro los negocios que estÃ¡n en el rubro con la mayor cantidad
businesses_in_max_line = nuevo_df[nuevo_df["line"] == max_line]

# Encuentro los 5 negocios con mayor cantidad de presencia en el rubro 
top_5_businesses = businesses_in_max_line["name"].value_counts().nlargest(5)

# Creao el grÃ¡fico de barras para mostrar los 5 rubros con mayor cantidad de presencia
plt.figure(figsize=(12, 6))
sns.barplot(x=top_5_businesses.values, y=top_5_businesses.index, palette='magma')
plt.xlabel('Cantidad de Negocios')
plt.ylabel('Nombre del Negocio')
plt.title(f'Top 5 Negocios en {max_line}', fontsize=14)
plt.show()
```

```python
# Encuentra el rubro con la segunda mayor cantidad de negocios
segundo_max_line = nuevo_df["line"].value_counts().index[1]  # El Ã­ndice 1 representa el segundo rubro mÃ¡s grande

# Filtra los negocios que estÃ¡n en el segundo rubro con la mayor cantidad
businesses_in_segundo_max_line = nuevo_df[nuevo_df["line"] == segundo_max_line]

# Encuentra los 5 negocios con mayor cantidad de presencia en el segundo rubro
top_5_businesses_segundo = businesses_in_segundo_max_line["name"].value_counts().nlargest(5)

# Crea el grÃ¡fico de barras para mostrar los 5 rubros con mayor cantidad de presencia
plt.figure(figsize=(12, 6))
sns.barplot(x=top_5_businesses_segundo.values, y=top_5_businesses_segundo.index, palette='magma')
plt.xlabel('Cantidad de Negocios')
plt.ylabel('Nombre del Negocio')
plt.title(f'Top 5 Negocios en {segundo_max_line}', fontsize=14)
plt.show()
```
